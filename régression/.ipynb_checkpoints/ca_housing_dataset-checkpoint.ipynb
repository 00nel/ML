# Regression
## Imports

```python
#manipulation de données
import pandas as pd
import numpy as np
#visualisation de données
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

#outils ML 
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split 

#modèles ML 
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

#métriques 
from sklearn.metrics import (mean_absolute_error, mean_squared_error, r2_score)
```

## Chargement et préparation des données 

```python
#chargement du dataset 
bunch = fetch_california_housing()
#df pour les features
X = pd.DataFrame(bunch.data, columns=bunch.feature_names)
#series pour la target 
y= pd.Series(bunch.target, name='MedHouseVal')
# Fusion des features et de la target dans un seul DataFrame
df = pd.concat([X, y], axis=1)
```

## Exploration des données 

```python
df.head()
```

```python
msno.matrix(df)
```

Il n'y a pas de valeurs manquantes dans ce df 
## Séparation train/test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=42)
```

## Modèle de regression

```python
def regression(X_train, X_test, y_train, y_test):
    """
    Fonction pour entrainer et évaluer plusieurs modèles de régression
    """
    #dict contenant les modèles 
    model = {"linear_reg" : LinearRegression(),
             "ridge" : Ridge(),
             "decision_tree" : DecisionTreeRegressor(),
             "random_forest" : RandomForestRegressor(),
             "gradient_boost" : GradientBoostingRegressor()
            }
    
    for a, b in model.items(): #boucler sur model
        print(f'------{a}------') #afficher le nom du modèle 
        b.fit(X_train, y_train) #compiler le modèle 
        y_pred = b.predict(X_test) #prédiction 

        #métriques d'évaluations
        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        rmse = np.sqrt(mse)
        r2 = r2_score(y_test, y_pred)
        print("----------REGRESSION METRICS----------")
        print(f"MAE   (Mean Absolute Error)          : {mae:.4f}")
        print(f"MSE   (Mean Squared Error)           : {mse:.4f}")
        print(f"RMSE  (Root Mean Squared Error)      : {rmse:.4f}")
        print(f"R²    (Coefficient de détermination) : {r2:.4f}")

        #visualisations des métriques 
        fig, axes = plt.subplots(1, 2, figsize=(15, 5))
        
        axes[0].scatter(y_test, y_pred, alpha=0.6, s=100, edgecolors='k')
        axes[0].plot([y_test.min(), y_test.max()], 
                     [y_test.min(), y_test.max()], 
                     'r--', lw=2, label='Prédiction parfaite')
        axes[0].set_xlabel('Valeur du bien', fontsize=12)
        axes[0].set_ylabel('Valeur prédite', fontsize=12)
        axes[0].set_title('Prédictions vs Valeurs Réelles', fontsize=14, fontweight='bold')
        axes[0].legend()
        axes[0].grid(True, alpha=0.3)
        
        erreurs = y_test - y_pred
        axes[1].hist(erreurs, bins=20, edgecolor='black', alpha=0.7)
        axes[1].axvline(x=0, color='red', linestyle='--', linewidth=2, label='Erreur nulle')
        axes[1].set_xlabel('Erreur de prédiction', fontsize=12)
        axes[1].set_ylabel('Fréquence', fontsize=12)
        axes[1].set_title('Distribution des Erreurs', fontsize=14, fontweight='bold')
        axes[1].legend()
        axes[1].grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()     
```

```python
regression(X_train, X_test, y_train, y_test)
```

## Analyse des résultats 
Dans ce dataset, prix réel d'un bien = MedHouseVal * 100 000
-> Les erreurs doivent être lues en valeur économique réelle.
### Régression Linéaire 
* En moyenne, le modèle se trompe de plus ou moins 53 000 USD (MAE : 0.5272) : l'erreur est trop élevée
* modèle raisonnable pas précis
* Erreur moyenne pondérée d'environ 73 000 USD, c'est trop élevée
* Le modèle explique que 60% de la variance des prix
* modèle stable mais erreurs trop élevée pour une prédiction fiable
### Ridge 
* résultat comparable à la régression linéaire
* tout comme pour la régression linéaire: si les features sont non ou peu corrélées entres elles, les modèles linaires ne sont pas adaptés
### Decision Tree 
* en moyenne, le modèle se trompe de 47 000 USD
* Erreur moyenne pondérée d'environ 73 000 USD
-> en général le modèle se trompe de 47 000 à 50 000 USD
* explique seulement 59% de la variance
* Le modèle améliore l’erreur moyenne absolue, indiquant une meilleure précision sur les prédictions courantes.
* En revanche, la dégradation du RMSE et du R² révèle la présence d’erreurs importantes ponctuelles et une moindre capacité de généralisation.
* Ces résultats sont cohérents avec la nature flexible mais instable des arbres de décision.
### Random Forest 
* le modele explique 80% de la variance : c'est satisfaisant 
* en moyenne, le modèle se trompe de 33 000 USD : correcte 
* erreur moyenne pondéré de 50000 USD, le modèle est plus stable
### Gradient Boosting 
* Le modèle explique environ 78% de la variance : c'est correct
* erreur mouenne de 37 000 à 53 000 USD : toujours trop élevée

## Conclusion 
Random Forest est le plus adapté pour cette problématique de régression 
NotaBene: Tout les modèles ont été utilisé avec leurs hyperparamêtres par défaut. GradientBoosting, avec les réglages optimales, aurait été plus performant que RandomForest
